**Question 19** [3 pts]: Expliquer en trois ou quatre phrases un algorithme pouvant vérifier étant donné un jeu de données `D`, une liste `QIDs` de quasi-identifiants et un paramètre `k` que `D` est **k-anonymisé**

D'abord, on crée une clé (index) avec les QIDs correspondant à chaque rangée.

Ensuite, on groupe le jeu de donnée en fonction des QIDs.

Puis, on compte le nombre d'éléments (rangées) dans chaque groupe.

Le nombre k correspondra au nombre de rangé du groupe,

**Question 20** [5 pts]: Proposer une implémentation de cet algorithme qui devra prendre en entrée un dataframe `df`, une liste de quasi-identifiant `QIDs` et le paramètre de k-anonymat `k`.

```python
def is_k_anonymat(df, QIDs, k):
  df = df.copy()
  df["QIDS"] = df[QIDs].apply(lambda qids : tuple(qids), axis=1)
  df_grouped = df.groupby("QIDS").size()
  df_grouped = all(df_grouped >= k)
  return df_grouped
```

**Question 22** [3 pts]: Donner la liste des pairs de quasi-identifiants pour lesquels `adult_df` satisfait le 2-anonymat.

Le **k-anonymat** est un critère de confidentialité qui stipule que pour chaque combinaison de valeurs des quasi-identifiants, il doit y avoir au moins k enregistrements dans le jeu de données. Dans ce cas, k = 2

```python
from itertools import combinations
def get_qids_for_k_anonymat(df,len_qids,k):

  df = df.copy()
  columns = df.columns.tolist()

  for combo in combinations(columns, len_qids):
    result = is_k_anonymat(df, list(combo), k)
    if result :
        return list(combo)
  return []

```

```python
list_2_anonymat = get_qids_for_k_anonymat(adult_df,2,2)
list_2_anonymat
```

**Question 25** [3 pts]: Expliquer en trois ou quatre phrases comment l'algorithme précédent pourrait être modifié pour vérifier qu'un jeu de données anonymisé satisfait la **l-diversité** pour un attribut sensible `s`.

On créer une colonne qui groupe les QIDs.

On vérifie la condition de k-anonymat avec l'algorithme implémenté précédemment

On vérifie si dans les groupes crées, on retrouve l'attribut sensible répétés au moins l fois.

Si les deux conditions précédentes sont vérifiés alors le jeu de donnée satisfat la l-diversité.



**Question 26** [5 pts]: Proposer une implémentation de cet algorithme qui devra prendre en entrée un dataframe `df`, une liste de quasi-identifiant `QIDs`, un attribut sensible `s`, le paramètre de k-anonymat `k` et le paramètre de l-diversité `l`.

```python
def is_l_diverse(df,QIDs, s, k, l):
  df = df.copy()
  df["QIDS"] = df[QIDs].apply(lambda qids : tuple(qids), axis=1)
  df_grouped = df.groupby("QIDS")

  columns = df.columns.tolist()


  for o, group in df_grouped:
    if group[s].nunique() >= l and is_k_anonymat(df,QIDs,k):
      return True

  return False
```

**Question 28** [5 pts]:  Étant donné le jeu de données `adult_df` qui satisfait le 3-anonymat pour QIDs = `["Sex", "Race"]`, parmi les attributs `["Age", "Workclass", "Relationship", "Country"]`, lesquels pourraient permettre une attaque d'homogénéité s'ils étaient publiés comme attributs sensibles?

```python
def is_l_homogeneous(df,QIDs, s, k):
  df = df.copy()
  df["QIDS"] = df[QIDs].apply(lambda qids : tuple(qids), axis=1)
  df_grouped = df.groupby("QIDS")

  columns = df.columns.tolist()
  for o, group in df_grouped:
    if group[s].nunique() == 1 and is_k_anonymat(df,QIDs,k):
      return True
  return False

```

**Question 29** [5 pts]: Proposer un algorithme pour généraliser l'attribut `Zip` en transformant les `n` derniers chiffres en `0`. Votre algorithme devra retourner la version transformée du jeu de données.

```python
def generalize_zip(df,n):
  df = df.copy()
  df["Zip"] = df["Zip"].apply(lambda zip: str(zip)[:-n] + '0'*n)
  return df
```
